```python
import numpy as np
import pandas as pd
from dateutil.parser import parse
```

```python
df = pd.read_csv('parliament.csv')
```

```python
relevant_features = ['BillNumberOfPages', 'BillType', 'DaysUntilRoyalAssent', 'ParliamentNumber', 'SessionNumber', 'PrimeMinisterName', 'PrimeMinisterParty', 'SponsorGender', 'SponsorParty', 'SponsorTitle', 'BillIntroducedDate']
```

```python
relevant_features
```

```python
features_raw = df[relevant_features]
```

```python
from datetime import datetime
```

```python
year = pd.to_datetime(features_raw.BillIntroducedDate).dt.year
```

```python
first_day_of_year = pd.Series([datetime(y, 1, 1) for y in year])
```

```python
features_raw.loc[:,'BillIntroducedDate'] = pd.to_datetime(features_raw.loc[:, 'BillIntroducedDate'])
```

```python
features_raw['BillDayOfYear'] = (features_raw['BillIntroducedDate'] - first_day_of_year).dt.days
```

```python
features_raw['BillMonthOfYear'] = ((features_raw.BillIntroducedDate).dt.month - first_day_of_year.dt.month) + 1
```

```python
features_raw.drop('BillIntroducedDate', axis=1, inplace=True)
```

```python
features_raw['IsSponsorInPower'] = features_raw['PrimeMinisterParty'] == features_raw['SponsorParty']
```

```python
features_raw['SponsorTitle'][features_raw['SponsorTitle'].isin((features_raw['SponsorTitle'].value_counts()[features_raw['SponsorTitle'].value_counts() < 8]).index)] = 'Other'
```

```python
features_raw['SponsorTitle'].value_counts()
```

```python
features_raw.ParliamentNumber = features_raw.ParliamentNumber.astype('category')
features_raw.SessionNumber = features_raw.SessionNumber.astype('category')
features_raw.BillMonthOfYear = features_raw.BillMonthOfYear.astype('category')
```
